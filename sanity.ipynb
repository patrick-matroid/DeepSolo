```python
import copy
t_parseq = copy.deepcopy(t_)
t_parseq['images'] = []
from glob import glob
from PIL import Image

import os.path as osp
for img in glob('datasets/ic15/test_images/*[.jpg, .png]'):
    img_obj = Image.open(img)
    template = copy.deepcopy(t_['images'][0])
    template['file_name'] = osp.basename(img)
    template['id'] = osp.splitext(img.split('_')[-1])[0]
    template['width'], template['height'] = img_obj.size
    t_parseq['images'].append(template)
```

```python
with open('datasets/ic15/test.json', 'w') as f:
    json.dump(t_parseq, f)
```
